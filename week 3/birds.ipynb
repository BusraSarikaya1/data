```python
# Visualize birds wingspan
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
wingspan = birds.MaxWingspan # Kuş veri kümesindeki "MaxWingspan" sütununu seçer ve "wingspan" adlı bir Seri (Pandas Serisi) oluşturur.
wingspan.plot()  #Oluşturulan Seri üzerinde çizgi grafiğini oluşturur.
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.DataFrame(birds)

# Plot the histogram thanks to the distplot function
sns.distplot( a=df["MaxLength"], hist=True, kde=True, rug=False )
```

```python
# Check outliers      #yani yukarıdaki grafikte max kanat açıklığı değerine sahip kuşun ismi Bald eagle mış diğeri ise Prairie falcon muş.
birds[birds["MaxWingspan"] > 500]
```

```python
# Label outliers
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both', which='both', labelbottom=False, bottom=False)

for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    plt.plot(x, y, "o", color="steelblue")
    if birds['MaxWingspan'][i] > 500:
        plt.text(x, y * (1 - 0.05), birds['Name'][i], fontsize=10,)# kanat açıklığı 500 den büyükse kuş ismini etiketler
    
plt.show()
```

```python
# 1- Remove outliers in plot

plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    if birds['Name'][i] not in ['Bald eagle', 'Prairie falcon']:
        plt.plot(x, y, "o",color="steelblue")
plt.show()
```
